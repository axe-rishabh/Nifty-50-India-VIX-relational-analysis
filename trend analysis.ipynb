```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nf = pd.read_csv('nifty_2020_2021.csv')
nf.head()
```

```python
nf.dtypes
```

```python
nf.tail()
```

```python
nf['Close'].pct_change()
```

```python
nf['change'] = nf['Close'].diff().round(2)
```

```python
nf['percent_change'] = (nf['Close'].pct_change() * 100).round(2)
```

```python
nf['Date'] = nf['Date'].astype('datetime64[ns]')
```

```python
nf.head()
```

```python
vix = pd.read_csv('ind_vix_2020_2021.csv')
vix.head()
```

```python
vix.drop('Vol.', axis=1, inplace=True)
```

```python
vix.head(5)
```

```python
vix.columns =['date','close', 'open', 'high', 'low', 'percent_change']
```

```python
vix['date'] = vix['date'].astype('datetime64[ns]')
```

```python
vix.head()
```

```python
vix = vix[::-1]
```

```python
vix = vix.reset_index(drop=True)
vix.head(5)
```

```python
vix['percent_change'] = vix['percent_change'].str.replace("%","")
vix['percent_change'] = vix['percent_change'].astype('float')
```

```python
vix.head(5)
```

```python
nf['percent_change'].corr(vix['percent_change'])
```

```python
corr_matrix = nf.corr(method='pearson')
corr_matrix
```

```python
fig1 = plt.figure(figsize=(10,10))
sns.heatmap(corr_matrix,xticklabels=corr_matrix.columns,yticklabels=corr_matrix.columns,
              cmap = 'YlGnBu',annot = True, linewidth=0.5)
print('Correlation between Stocks in you portfolio')
plt.show(fig1)
```

```python
vol_avg = [np.mean(nf['Volume'][n_row:])] * len(nf[n_row:])
```

```python
plt.figure(figsize=(15,6))
plt.plot(nf['Date'][n_row:],nf['Volume'][n_row:],marker = 'o',label='Volume')
plt.plot(nf['Date'][n_row:],vol_avg,label = 'average')
plt.title('NIFTY')
plt.legend()
```

```python
nf_per_avg = [np.mean(nf['percent_change'][n_row:])] * len(nf[n_row:])
nf_per_avg = [np.mean(vix['percent_change'][n_row:])] * len(nf[n_row:])

plt.figure(figsize=(15,8))

plt.plot(vix[n_row:]['date'],vix[n_row:]['percent_change'],marker='o',label = 'Vix',color = 'black')
plt.plot(nf[n_row:]['Date'],nf[n_row:]['percent_change'],marker='o', label = 'nifty',color = 'green')
plt.plot(nf[n_row:]['Date'], nf_per_avg,label = 'Nifty Percent Average')
plt.plot(vix[n_row:]['date'], nf_per_avg,label = 'VIX Percent Average')
plt.title('Nifty Vix PERCENTAGE CHANGE comparison')
plt.legend()
plt.ylim(-10.0,30.0)
```
